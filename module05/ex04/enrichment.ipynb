```python
import pandas as pd
import numpy as np
import requests

path = '/content/drive/My Drive/data/auto.json' 
```

```python
df = pd.read_json(path)
pd.options.display.float_format = '${:,.2f}'.format
df
```

```python
df_new = df.sample(n=200, random_state=21)
df_new
```

```python
concat_rows = pd.concat([df, df_new], ignore_index=True)
concat_rows
```

```python
np.random.seed(21)
list = np.random.randint(1980, 2020, size=925)
concat_rows['Year'] = list
fines = concat_rows
fines
```

```python
path2='/content/drive/My Drive/data/surname.json' 
```

```python
df_surname = pd.read_json(path2)
df_surname.rename(columns=df_surname.iloc[0], inplace=True)
df_surname.sort_values(by=['COUNT'], ascending=False, inplace=True)
df_surname
```

```python
car_numb = fines.drop_duplicates('CarNumber')['CarNumber']
car_numb = car_numb.reset_index(drop=True)
car_numb
```

```python
owner_name = df_surname['NAME'].sample(n=531, random_state=21, replace=True, ignore_index=True)
owner_name
```

```python
owners = pd.DataFrame(data=car_numb)
owners['SURNAME'] = owner_name
owners
```

```python
additional = fines.sample(n=5, random_state=21)
additional
```

```python
fines = pd.concat([fines, additional], ignore_index=True)
fines
```

```python
owners = owners.drop(owners.tail(20).index)
additional2 = owners.sample(n=3, random_state=21)
owners = pd.concat([owners, additional2], ignore_index=True)
owners
```

```python
new1 = pd.merge(fines, owners, on='CarNumber', how='inner')
new1
```

```python
new2 = pd.merge(fines, owners, on='CarNumber', how='outer')
new2
```

```python
new3 = pd.merge(fines, owners, on='CarNumber', how='left')
new3
```

```python
new4 = pd.merge(fines, owners, on='CarNumber', how='right')
new4
```

```python
pd.pivot_table(fines, columns='Year', values='Fines', index=['Make', 'Model'], aggfunc={'Fines': np.sum})
```

```python
fines.to_csv('/content/drive/My Drive/data/fines.csv', index=False)
```
