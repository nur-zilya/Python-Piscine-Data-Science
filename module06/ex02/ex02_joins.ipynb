```python
import pandas as pd
import sqlite3
```

```python
data = sqlite3.connect('/content/drive/My Drive/data/checking-logs.sqlite.sqlite')
```

```python
query = "SELECT checker.uid, checker.labname, \
checker.timestamp AS 'first_commit_ts', \
pageviews.datetime as 'first_view_ts' \
from checker \
LEFT JOIN pageviews ON checker.uid=pageviews.uid \
WHERE checker.status='ready' AND checker.numtrials=1 \
AND checker.labname IN ('laba04', 'laba04s', 'laba05', 'laba06', 'laba06s', 'project1') \
AND checker.uid LIKE 'user_%' \
AND (pageviews.datetime in \
(SELECT info_time from \
(SELECT uid, MIN(datetime) as info_time \
from pageviews group by uid)) or pageviews.datetime is NULL)"
datamart = pd.io.sql.read_sql(query, data, parse_dates=['first_commit_ts', 'first_view_ts'])
datamart
```

```python
test = datamart[datamart['first_view_ts'].notnull()]
test
```

```python
control = datamart[datamart['first_view_ts'].isnull()]
control
```

```python
average = test['first_view_ts'].mean()
control.fillna(average, inplace=True)
control
```

```python
test.to_sql('./test', data)
control.to_sql('./control', data)
```
